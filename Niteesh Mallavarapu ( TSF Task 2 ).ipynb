# CONCLUSION:

*) Standard Class ShipMode transported more quantities than any other ShipMode and gained profits. Hence, we have to increase the frequency of Standard Class ShipMode.

*) Sales should focus more on Consumer Segment than other segments for more profit.

*) SameDay ShipMode received transported with major discount offers which can effect profits. To avoid this, discounts has to be limited to certain range.

*) Copiers have shown best profits in all regions.Additional profits can be obtained by increasing sales in regions.

*) Technology dominated Office Supplies Category and Furniture in terms of Sales.
